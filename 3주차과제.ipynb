# BMI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import koreanize_matplotlib

pd.options.display.float_format = '{:.2f}'.format

plt.rcParams['figure.figsize'] = [8,5]
```

```python
np.random.seed(42)

data_male = {
    'weight' : np.random.normal(75,10,50),
    'height' : np.random.normal(170,10,50),
    'gender' : '남성'
}
df_male = pd.DataFrame(data_male)

data_female = {
    'weight' : np.random.normal(50,5,50),
    'height' : np.random.normal(160,5,50),
    'gender' : '여성'
}
df_female = pd.DataFrame(data_female)

df = pd.concat([df_male, df_female ], axis=0)

df
```

```python
def calBMI(weight, height):
    return weight/(height/100)**2

df['BMI'] = calBMI(df['weight'],df['height'])

df.head()
```

```python
def BMIToCategory(bmi):
    if bmi < 18.5:
        return '저체중'
    elif 18.5 <=bmi <23:
        return '정상'
    elif 23 <= bmi < 25:
        return '과체중'
    elif 25 <= bmi < 30:
        return '경도비만'
    elif 30 <= bmi < 35:
        return '중등도 비만'
    else:
        return '고도비만'
```

```python
df['비만도'] = df['BMI'].apply(BMIToCategory)

df.head()
```

```python
desc_stats = df.describe()

print("기술통계:\n", desc_stats)
```

```python
sns.histplot(
    df['BMI'],
    kde=True
    )

plt.title('BMI분포')
plt.xlabel('BMI')
plt.ylabel('빈도수')
plt.show()
```

```python
sns.histplot(
    data=df,
    x = 'BMI',
    kde=True,
    hue='gender'
    )

plt.title('BMI분포')
plt.xlabel('BMI')
plt.ylabel('빈도수')
plt.show()
```

```python
# 비만도 실제 값 확인
print(df['비만도'].unique())
```

# 혈압

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import koreanize_matplotlib

pd.options.display.float_format = '{:.2f}'.format

plt.rcParams['figure.figsize'] = [8,5]
```

```python
np.random.seed(42)

data = {
    'SBP' : np.random.normal(120,15,100)
}

df = pd.DataFrame(data)

df.head()
```

```python
df['DBP'] = df['SBP'] - np.random.randint(30,50,100)

df.head()
```

```python
desc_stats = df.describe()

print("기술통계:\n", desc_stats)
```

```python
fig, ax = plt.subplots(1,2)

sns.histplot(
    df['SBP'],
    kde=True,
    ax=ax[0]
)

ax[0].set_title('수축기 혈압 분포')
ax[0].set_xlabel('mmHg')
ax[0].set_ylabel('빈도수')

sns.histplot(
    df['DBP'],
    kde=True,
    ax=ax[1]
)

ax[1].set_title('이완기 혈압 분포')
ax[1].set_xlabel('mmHg')
ax[1].set_ylabel('빈도수')

plt.tight_layout()

plt.show()
```

```python
corr = df['SBP'].corr(df['DBP'])

print("수축기 혈압과 이완기 혈압의 상관관계:",corr)
```

```python
sns.scatterplot(
    data=df,
    x='SBP',
    y='DBP'
)
plt.title('수축기 혈압과 이완기 혈압의 상관관계')
plt.xlabel('수축기 혈압(mmHg)')
plt.ylabel('이완기 혈압(mmHg)')
plt.show()
```

## 콜레스트롤 수치

[총 콜레스테롤 기준]
정상: 200 mg/dL 미만

경계(주의): 200 ~ 239 mg/dL

이상(높음): 240 mg/dL 이상

```python
# 필요한 라이브럴리 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 폰트
import koreanize_matplotlib

# pandas 데이터 값 실수, 소수점 둘째까지
pd.options.display.float_format = '{:.2f}'.format

# 그래프 크기설정
plt.rcParams['figure.figsize'] = [8,5]
```

```python
#랜덤 시드 고정
np.random.seed(42)

data = {
    'TC' : np.random.normal(200,30,100), # 평균 200mg/dL, 표준편차 30mg/dL
    'age' :  np.random.randint(18,80,100), # 18~88세사이
    'gender' :  np.random.choice(['Male', 'Female'],100) #성별
    }

df = pd.DataFrame(data)

df.info()
```

```python
#총콜레스트롤양을 이용한 콜레스트롤수치 분류함수
def classifyCholesterol(tc):
    if tc < 200:
        return '정상'
    elif 200 <= tc < 239:
        return '경계(주의)'
    else:
        return '이상(높음)'
```

```python
#함수 정상작동 확인
classifyCholesterol(210)
```

```python
#콜레스트롤 수치 분류
#콜레스트롤 수치 컬럼에 값에따라 분류결과 작성
#lambda x 로 콜레스트롤 수치 분류함수 적용
df['콜레스트롤 수치'] = df['TC'].apply(lambda x: classifyCholesterol(x))

#데이터 프레임 출력
df.head()
```
